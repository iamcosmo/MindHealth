```python
# optimizing the gender column
gender = df['Gender'].str.lower()
gender = df['Gender'].unique()
gender
```

```python
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]

trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means", "A little about you", "p"]           

female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

column = 'Gender'
for value in df[column]:

    if str.lower(value) in male_str:
        df['Gender'].replace(to_replace=value, value='male', inplace=True)
        
    if str.lower(value) in female_str:
        df['Gender'].replace(to_replace=value, value='female', inplace=True)
        
    if str.lower(value) in trans_str:
        df['Gender'].replace(to_replace=value, value='trans', inplace=True)
    
# Since this one value 'A little value is not replacing' so we can replace it separately.
old = 'A little about you'
new = 'trans'
cond = df[column] == old
df.loc[cond,column] = new

print(df['Gender'].unique())
```
